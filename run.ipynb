## Clean up `/logs` and `/models/model_head`

```python
# Clear all files in /logs
import os
folder_path = 'logs'

# Loop through the directory
for entry in os.scandir(folder_path):
    if entry.is_file():
        try:
            os.remove(entry.path)
            print(f"Deleted: {entry.path}")
        except Exception as e:
            print(f"Error deleting {entry.path}: {e}")
```

```python
# Clear all files in /models/model_head
import os
folder_path = 'models/model_head'

# Loop through the directory
for entry in os.scandir(folder_path):
    if entry.is_file():
        try:
            os.remove(entry.path)
            print(f"Deleted: {entry.path}")
        except Exception as e:
            print(f"Error deleting {entry.path}: {e}")
```

## Save Results to persisted stored

```python
# Download results directory as zip
import shutil
from google.colab import files

folder_to_save = 'results/'
zip_name = 'model_results' # Do not add .zip here, shutil does it

# 1. Zip the folder
shutil.make_archive(zip_name, 'zip', folder_to_save)
print(f"Created {zip_name}.zip")

# 2. Trigger the browser download
files.download(f"{zip_name}.zip")
```
